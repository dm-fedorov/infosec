```python
%cd -q ..

import graphviz

def make_graph(node_label=None, graph_name=None, **node_kwargs):
    result = graphviz.Digraph(name=graph_name)
    result.node('A', label=node_label, **node_kwargs)
    print(result)
    return result
```

```python
make_graph(node_label='"')
```

```python
make_graph(node_label=r'\\')
```

```python
make_graph(node_label=r'\\"')
```
